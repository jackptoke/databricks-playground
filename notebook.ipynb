```python
# Create the Spark session via Databricks Connect.
# `spark` is auto-injected in the Databricks web UI, but NOT when running
# locally in VSCode — so we build it here. The Databricks VSCode extension
# supplies the cluster/host/auth via environment variables.
from databricks.connect import DatabricksSession

spark = DatabricksSession.builder.getOrCreate()
```

```python
spark.sql("""
SELECT *
FROM nyctaxi.gold.daily_trips_summary
WHERE pickup_date = '2024-01-02'
""")
```
