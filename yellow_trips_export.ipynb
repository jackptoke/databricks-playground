```python
from pyspark.sql.functions import date_format

most_recent_run = dbutils.widgets.get("most_recent_run")
df = spark.sql("SELECT * FROM nyctaxi.silver.trips WHERE processed_timestamp >= :most_recent_run", args={"most_recent_run": most_recent_run})
df = df.withColumn("year_month", date_format("pickup_datetime", "yyyy-MM"))

output_path = "abfss://nyctaxi-yellow@nyctaxistorageaus2026.dfs.core.windows.net/yellow_trips_export/"
df.write \
  .mode("append") \
  .option("header", "true") \
  .option("path", output_path) \
  .format("csv") \
  .partitionBy("vendor", "year_month") \
  .saveAsTable("nyctaxi.export.yellow_trips_export")
```
